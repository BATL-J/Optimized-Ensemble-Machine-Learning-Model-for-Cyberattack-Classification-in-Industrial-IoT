# 1.FPR Calculation

## 1.1 DT FPR



```python
# precision, recall, and support values for each class (0-14)
classes = list(range(15))
precision = [0.96, 0.90, 1.00, 1.00, 1.00, 0.54, 1.00, 1.00, 1.00, 0.98, 0.95, 0.86, 0.96, 0.98, 0.90]
recall =    [0.95, 0.91, 1.00, 1.00, 1.00, 0.85, 1.00, 1.00, 1.00, 0.93, 0.95, 0.97, 0.84, 0.96, 0.90]
support =   [2082, 2101, 2781, 2012, 2869, 159, 77, 4872, 1978, 1707, 1971, 2022, 2012, 2076, 1934]

# Total number of samples
total_samples = sum(support)

# Calculate FPR per class
fpr_per_class = []
for i in range(len(classes)):
    TP = recall[i] * support[i]
    FN = support[i] - TP
    FP = TP / precision[i] - TP
    TN = total_samples - support[i] - FP
    FPR = FP / (FP + TN)
    fpr_per_class.append(FPR)

# Calculate macro average FPR
macro_avg_fpr = sum(fpr_per_class) / len(fpr_per_class)

# Show FPRs per class and macro average
for idx, fpr in enumerate(fpr_per_class):
    print(f"Class {idx} FPR: {fpr:.4f}")

print(f"\nMacro Average FPR: {macro_avg_fpr:.6f}")
```

## 1.7 AdaBoost FPR

```python
# precision, recall, and support values for each class (0-14)
classes = list(range(15))
precision = [0.00 , 0.00 , 1.00, 0.00 , 1.00, 0.00 , 0.00 , 1.00 , 1.00 , 0.05 , 0.24 , 1.00 , 0.34 , 0.00 , 0.32 ]
recall =    [0.00 , 0.00 , 1.00 , 0.00 , 1.00 , 0.00 , 0.00 , 1.00 , 0.00 , 0.00 , 1.00 , 0.00 , 1.00 , 0.00 , 1.00]
support =   [1956 , 2169 , 2835 , 2069 , 2916 , 162 , 71 , 4818 , 1970 , 1781 , 1920 , 2066 , 2053 , 1944 , 1922]

# Total number of samples
total_samples = sum(support)

# Calculate FPR per class
fpr_per_class = []
for i in range(len(classes)):
    TP = recall[i] * support[i]
    FN = support[i] - TP

    if precision[i] == 0:
        FP = 0  # or you could use None or skip these classes
    else:
        FP = TP / precision[i] - TP

    TN = total_samples - support[i] - FP
    FPR = FP / (FP + TN) if (FP + TN) != 0 else 0
    fpr_per_class.append(FPR)

# Calculate macro average FPR
macro_avg_fpr = sum(fpr_per_class) / len(fpr_per_class)

# Show FPRs per class and macro average
for idx, fpr in enumerate(fpr_per_class):
    print(f"Class {idx} FPR: {fpr:.4f}")

print(f"\nMacro Average FPR: {macro_avg_fpr:.6f}")
```

```python
# precision, recall, and support values for each class (0-14)
classes = list(range(9))
precision = [1.00 , 0.99 , 1.00 , 1.00 , 1.00 , 1.00 , 0.67 , 1.00 ,1.00]
recall =    [1.00 , 0.90 , 0.99 , 1.00 , 1.00 , 0.50 , 0.29 , 1.00 , 1.00]
support =   [412651 , 391 , 1830 , 25603 , 38747, 2, 7, 18164 , 429 ]

# Total number of samples
total_samples = sum(support)
# Calculate FPR per class
fpr_per_class = []
for i in range(len(classes)):
    TP = recall[i] * support[i]
    FN = support[i] - TP
    FP = TP / precision[i] - TP
    TN = total_samples - support[i] - FP
    FPR = FP / (FP + TN)
    fpr_per_class.append(FPR)

# Calculate macro average FPR
macro_avg_fpr = sum(fpr_per_class) / len(fpr_per_class)

# Show FPRs per class and macro average
for idx, fpr in enumerate(fpr_per_class):
    print(f"Class {idx} FPR: {fpr:.4f}")

print(f"\nMacro Average FPR: {macro_avg_fpr:.6f}")
```
